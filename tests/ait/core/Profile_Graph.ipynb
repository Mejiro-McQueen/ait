```python
#import seaborn as sns
import pandas as pd
from pathlib import Path
import os
```

```python
dfs = (pd.read_csv(i) for i in Path('./').glob('*.csv') if os.path.getsize(i))
for df in dfs:
    df['unit_time'] = df.apply(lambda row: row.time_s / row.msg_len, axis=1)
    df['time_normal'] = (df.time_s - df.time_s.min()) / (df.time_s.max() - df.time_s.min())
    df.plot(y='time_s', label=df.process[0])
```
